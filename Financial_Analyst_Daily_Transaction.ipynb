## Identify Missing Values

### Subtask:
Determine the extent of missing values across all columns in the DataFrame to inform handling strategies.


**Reasoning**:
To understand the extent of missing values, I will calculate both the count and percentage of missing values for each column in the DataFrame `df` and display them.



```python
missing_values_count = df.isnull().sum()
missing_values_percentage = (df.isnull().sum() / len(df)) * 100

missing_values_df = pd.DataFrame({
    'Missing Count': missing_values_count,
    'Missing Percentage': missing_values_percentage
})

print("Missing Values Report:")
print(missing_values_df[missing_values_df['Missing Count'] > 0].sort_values(by='Missing Percentage', ascending=False))
```

## Handle Missing 'Date' Values

### Subtask:
Drop rows where the 'Date' column is missing, as these entries are crucial for time-based analysis and cannot be reliably imputed without further context.


**Reasoning**:
To handle missing values in the 'Date' column, I will drop rows where 'Date' is NaN, as per the instructions, and then calculate how many rows were removed and display the new shape of the DataFrame.



```python
original_rows = df.shape[0]
df_cleaned = df.dropna(subset=['Date'])
removed_rows = original_rows - df_cleaned.shape[0]
df = df_cleaned.copy()

print(f"Number of rows removed due to missing 'Date' values: {removed_rows}")
print(f"New DataFrame shape after dropping rows with missing 'Date': {df.shape}")
```

## Handle Missing 'Subcategory' and 'Note' Values

### Subtask:
Fill missing values in the 'Subcategory' and 'Note' columns with a placeholder like 'Unknown' to retain as much data as possible while making the missing information explicit.


#### Reasoning:
As specified, I will replace all `NaN` values in both 'Subcategory' and 'Note' columns with the string 'Unknown'. This approach ensures that no data is lost while making it clear where information was originally missing.

**Reasoning**:
To fill missing values in 'Subcategory' and 'Note' columns with 'Unknown' and then display the missing value counts for these columns to verify the operation.



```python
df['Subcategory'].fillna('Unknown', inplace=True)
df['Note'].fillna('Unknown', inplace=True)

print("Missing values after filling 'Subcategory' and 'Note':")
print(df[['Subcategory', 'Note']].isnull().sum())
```

**Reasoning**:
The previous code generated a `FutureWarning` due to the use of `inplace=True`. To resolve this and align with best practices, I will remove `inplace=True` and reassign the results of the `fillna` operation back to the respective columns.



```python
df['Subcategory'] = df['Subcategory'].fillna('Unknown')
df['Note'] = df['Note'].fillna('Unknown')

print("Missing values after filling 'Subcategory' and 'Note':")
print(df[['Subcategory', 'Note']].isnull().sum())
```

## Calculate daily net income/expense

### Subtask:
Calculate the daily net income/expense (income minus expense) from the 'Amount' and 'Income/Expense' columns.


**Reasoning**:
To calculate the daily net income/expense, I will first create a new column 'Net_Amount'. This column will store the amount as negative if it's an 'Expense' or 'Transfer-Out', and positive if it's 'Income'.



```python
df['Net_Amount'] = df.apply(lambda row: -row['Amount'] if row['Income/Expense'] in ['Expense', 'Transfer-Out'] else row['Amount'], axis=1)
print("Created 'Net_Amount' column. Displaying first 5 rows with new column:")
display(df.head())
```

**Reasoning**:
Now that the 'Net_Amount' column is created, I will group the DataFrame by date and sum the 'Net_Amount' to get the daily net flow, storing it in `daily_net_flow`.



```python
daily_net_flow = df.groupby(df['Date'].dt.date)['Net_Amount'].sum()
print("Daily net income/expense calculated. Displaying first 5 days:")
display(daily_net_flow.head())
```

**Reasoning**:
To visualize the cumulative balance, I will calculate the cumulative sum of the `daily_net_flow` and then plot it as a time series to show the financial balance evolution over time.



```python
cumulative_balance = daily_net_flow.cumsum()

plt.figure(figsize=(14, 7))
plt.plot(cumulative_balance.index, cumulative_balance.values, marker='o', linestyle='-')
plt.title('Cumulative Balance Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Balance (INR)')
plt.grid(True)
plt.tight_layout()
plt.show()

print("Cumulative balance calculated and plotted.")
```

**Reasoning**:
To identify days with significant positive or negative net flow, I will calculate thresholds based on the mean and standard deviation of the `daily_net_flow`, specifically looking for days where the net flow is more than two standard deviations away from the mean.



```python
mean_net_flow = daily_net_flow.mean()
std_net_flow = daily_net_flow.std()

positive_threshold = mean_net_flow + 2 * std_net_flow
negative_threshold = mean_net_flow - 2 * std_net_flow

significant_positive_days = daily_net_flow[daily_net_flow > positive_threshold]
significant_negative_days = daily_net_flow[daily_net_flow < negative_threshold]

print(f"Mean Daily Net Flow: {mean_net_flow:.2f}")
print(f"Standard Deviation of Daily Net Flow: {std_net_flow:.2f}")
print(f"Significant Positive Net Flow Threshold (>2 Std Dev): {positive_threshold:.2f}")
print(f"Significant Negative Net Flow Threshold (<2 Std Dev): {negative_threshold:.2f}")

print("\nDays with Significant Positive Net Flow:")
display(significant_positive_days)

print("\nDays with Significant Negative Net Flow:")
display(significant_negative_days)
```
